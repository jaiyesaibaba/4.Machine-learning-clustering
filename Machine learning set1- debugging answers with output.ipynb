```python
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
```

```python
X_train = np.array([[1, 2], [3, np.nan], [5, 6]])
y_train = np.array([10, 20, 30])
```

```python
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
```

```python
model = LinearRegression()
model.fit(X_train_imputed, y_train_encoded)
```

```python
prediction = model.predict([[3, 4]])
print("Predicted Output:", prediction)
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression

imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)


model = LinearRegression()
model.fit(X_train_imputed, y_train_encoded)


print(model.predict(imputer.transform([[2, 3]])))

```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
import numpy as np

preprocessor = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),  # Replace NaN with mean values
    ('scaler', StandardScaler())
])


X_train_processed = preprocessor.fit_transform(X_train)


model.fit(X_train_processed, y_train)


X_test = np.array([[1, 150]])
X_test_processed = preprocessor.transform(X_test)

# Make prediction
print(model.predict(X_test_processed))
```
